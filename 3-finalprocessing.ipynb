```python


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df=pd.read_csv("../input/quoraquestionpair2dataprocessing/nlp_features_train.csv")
df.head()
```

```python
print(df.columns)
print(df.shape)
```

```python
df['question1']=df['question1'].apply(lambda x:str(x))
df['question2']=df['question2'].apply(lambda x:str(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
# merge texts
questions = list(df['question1']) + list(df['question2'])

tfidf = TfidfVectorizer(lowercase=False, )
tfidf.fit_transform(questions)
word2tfidf = dict(zip(tfidf.get_feature_names(), tfidf.idf_))
word2tfidf
```

We want to use glove + tfidf vect method , hence we multiply each word-vector with idf score

1. we take a sentence , and pass it to pretrained glove model
2. create a zero matrix of len(sent) * dim(i.e. 300)
3. for each word from the sentence find its corresponding value from the word2tfidf and multiply with vector(which we got from glove )
4. reapeat step 3 for each word, and finally add all of them and take mean value
5. store the mean value and repeat  step 1 to 4 for each senctence
6.

```python
df=df[:1000]
df
```

```python
import spacy
from tqdm import tqdm

nlp = spacy.load('en_core_web_lg')
vecs = []

#due to high computation time, we are using top 1k records
for qu in tqdm(list(df['question1'][:1000])):
    doc = nlp(qu)
    mean_vec = np.zeros([len(doc), len(doc[0].vector)])
    for word in doc:
        vec = word.vector
        try:
            idf = word2tfidf[str(word)]
        except:
            idf = 0        
        mean_vec += vec * idf
    mean_vec = mean_vec.mean(axis=0)
    vecs.append(mean_vec)
    
df['q1_feats_m'] = list(vecs)       
```

```python
vecs = []
for qu in tqdm(list(df['question2'])):
    doc = nlp(qu)                                         # doc and qu are same , but in doc.vector each word is represented by 300 dim 
    mean_vec = np.zeros([len(doc), len(doc[0].vector)])   # mean vector is zero vector of size length of sent * 300
    for word in doc:
        vec = word.vector                                 #vec =[, , , ,] list of size 300
        try:
            idf = word2tfidf[str(word)]                   # finding idf value from dict
        except:
            idf = 0        
        mean_vec += vec * idf
    mean_vec = mean_vec.mean(axis=0)
    vecs.append(mean_vec)
    
df['q2_feats_m'] = list(vecs)  
```
